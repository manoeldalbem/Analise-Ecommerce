```python
# Fazendo merge entre as tabelas. 
# As chaves utilizadas foram as colunas comuns entre todas as tabelas: Mês, Departamento, UF e Item.
agregado = financeiro.merge(vendas, on = ["Mes", "Departamento", "UF", "Item"]).merge(visitas, on = ["Mes", "Departamento", "UF", "Item"])
```

```python
def convert_p(dataset):
    """
    primeiro faz o replace da vírgula pelo ponto, uma vez que no sistema americano, ponto é que separa a unidade da parte fracionária.
    por fim transforma a string em float.
    
    input é uma coluna string.
    output é a coluna transformada em float
    """
    return float(dataset.replace(",","."))
```

```python
# colunas que são string a serem convertidas em float
colunas_a_transformar_em_float = ['Faturamento_Produto', 'Faturamento_Frete', 'Desconto_Produto', 'Desconto_Frete']

for coluna in colunas_a_transformar_em_float:
    agregado[coluna] = agregado[coluna].apply(convert_p)
```

```python
agregado['Departamento'].unique().tolist()
```

```python
def renomeando_maq_fot(dataframe_coluna):
    
    """
    conserta a ortografia quando a string for Maquinas Fotogr\xa0ficas
    input é a coluna ser reescrita
    output é a própria coluna corrigida
    """
    
    if dataframe_coluna == 'Maquinas Fotogr\xa0ficas':
        return 'Máquinas Fotográficas'
    else:
        return dataframe_coluna
```

```python
agregado['Departamento'] = agregado['Departamento'].apply(renomeando_maq_fot)
```

# Análise de Faturamento:

```python
# optei por fazer um sum de todas as colunas para analisar vários gráficos mais a frente.
departamento = agregado.groupby('Departamento').sum()
departamento
```

Cálculo do Faturamento Total, Custo Total, Desconto Total e Lucro, onde:

Faturamento Total = Faturamento_Produto + Faturamento_Frete.

Custo Total = Custo_Produto + Custo_Frete + Faturamento_Frete.

Desconto Total = Desconto_Produto + Desconto_Frete.

Lucro = Faturamento Total - Custo Total.

```python
# calculando faturamento total, custo total, desconto toal e lucro total
departamento['Faturamento_Total'] = departamento['Faturamento_Produto'] + departamento['Faturamento_Frete']
departamento['Custo_Total'] = departamento['Custo_Produto'] + departamento['Custo_Frete'] + departamento['Custo_Mkt']
departamento['Desconto_Total'] = departamento['Desconto_Produto'] + departamento['Desconto_Frete']
departamento['Lucro'] = departamento['Faturamento_Total'] - departamento['Custo_Total']
```

Cálculo dos Faturamentos, Custos e Lucro como proporção das vendas.

Cálculo da taxa de conversão de visitas em vendas.

```python
# calculo como proporção das vendas
departamento['Faturamento/Vendas'] = np.round(departamento['Faturamento_Total']/departamento['Vendas'], 2)
departamento['Custo/Vendas'] = np.round(departamento['Custo_Total']/departamento['Vendas'], 2)
departamento['Desconto/Vendas'] = np.round(departamento['Desconto_Total']/departamento['Vendas'], 2)
departamento['Lucro/Vendas'] = np.round(departamento['Lucro']/departamento['Vendas'], 2)
departamento['Desconto_Produto/Vendas'] = np.round(departamento['Desconto_Produto']/departamento['Vendas'], 2)
departamento['Desconto_Frete/Vendas'] = np.round(departamento['Desconto_Frete']/departamento['Vendas'], 2)
departamento['Custo_Mkt/Vendas'] = np.round(departamento['Custo_Mkt']/departamento['Vendas'], 2)
departamento['Vendas/Visitas'] = np.round(departamento['Vendas']/departamento['Visitas'], 2)
```

```python
data_mes['Faturamento_Total'] = data_mes['Faturamento_Produto'] + data_mes['Faturamento_Frete']
data_mes['Custo_Total'] = data_mes['Custo_Produto'] + data_mes['Custo_Frete'] + data_mes['Custo_Mkt']
data_mes['Desconto_Total'] = data_mes['Desconto_Produto'] + data_mes['Desconto_Frete']
data_mes['Lucro'] = data_mes['Faturamento_Total'] - data_mes['Custo_Total']
```

```python
data_mes
```

```python
def porcentagem_mes(dataframe_coluna, dataframe):
    """
    Faz o percentual por linha, ou seja, verifica pelo índice o quanto de percentual cada coluna de custo/investimento
    representa esse percentual.
    
    dataframe_coluna: coluna do dataframe a determinar o percentual.
    dataframe: dataframe a fim de fazer a soma pelo índice.
    
    output é o percentual da coluna que foi calculado"""
    
    pct = dataframe_coluna/(dataframe['Desconto_Produto']+dataframe['Desconto_Frete']+dataframe['Custo_Mkt'])
    return pct
```

Cálculo do percentual de desconto/investimento por mês:

```python
data_mes['Desconto_Produto_Pct'] = porcentagem_mes(data_mes['Desconto_Produto'], data_mes)
data_mes['Desconto_Frete_Pct'] = porcentagem_mes(data_mes['Desconto_Frete'], data_mes)
data_mes['Custo_Mkt_Pct'] = porcentagem_mes(data_mes['Custo_Mkt'], data_mes)
```

```python
def porcentagem_pelo_mesmo(dataframe_coluna):
    
    """
    Faz o percentual da coluna por ela mesma. Verifica em que período ficou mais concentrado cada coluna.
    
    input é a coluna a ser calculado o percentual.
    output é o percentual calculado de cada coluna.
    """
    
    pct = dataframe_coluna/dataframe_coluna.sum()
    return pct
```

```python
data_mes['Desconto_Produto_Pct_'] = porcentagem_pelo_mesmo(data_mes['Desconto_Produto'])
data_mes['Desconto_Frete_Pct_'] = porcentagem_pelo_mesmo(data_mes['Desconto_Frete'])
data_mes['Custo_Mkt_Pct_'] = porcentagem_pelo_mesmo(data_mes['Custo_Mkt'])
data_mes['Vendas_'] = porcentagem_pelo_mesmo(data_mes['Vendas'])
data_mes['Visitas_'] = porcentagem_pelo_mesmo(data_mes['Visitas'])
```

```python
data_mes['Vendas/Visitas'] = data_mes['Vendas']/data_mes['Visitas']
```
